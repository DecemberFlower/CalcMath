```python
import numpy as np
import math
from matplotlib import pyplot as plt
from numpy import linalg as LA
```

Имеем задачу

Обозначив z=y', имеем

```python
z = np.zeros(81)
y = np.zeros(81)
y[0] = 1
z[0] = -1
t = 0.01
for i in range(80):
    y[i+1] = y[i] + t * z[i]
    z[i+1] = z[i] + t * 10 * z[i] + 11 * t * y[i+1]
```

Для наглядности уменьшим отрезок, где решается задача до [0; 0.8]. Тогда разностная задача даёт решение вида:

```python
x = np.arange(0, 0.8+t, t)
plt.plot(x, y)
```

Общее решение диффура включает растущую и спадающую экспоненты. Начальные условия подобраны так, что остаётся только спадающая e^(-x). Тогда должно было получиться:

```python
x = np.arange(0, 0.8+t, t)
plt.plot(x, np.exp(-x))
```

Это вызванно тем, что предоставленные начальные условия не являются собственным вектором ограниченного решения разностной задачи, а значит там будет неограниченная компонента.
